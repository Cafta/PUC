## Exercício 1 - Resolvido:
Escreva uma função em Python que recebe uma string por parâmetro e remove um
carcter de acordo com númedo de sua posição (nth) e retorna o resultado na tela.
A string e o índice são recebidos via teclado.
Supõe que o usuário não insira um índice inválido

```python
def remove_char(str, n):
  fist_part = str[:n]
  last_part = str[n+1:]
  return fist_part + last_part
```

```python
word = input('Digite uma palavra: ')
nth = int(input('Digite o índice de remoção: '))
print(remove_char(word, nth))
```

## Exercício 2 - Praticar:
Escreva uma função em Python que recebe uma lista de palavras por parâmetro e
retorne a palavra mais longa da lista.
Assume-se que a lista de palavras já existe.

Exemplos:

lst = [‘php’, ‘python’, ‘inteligentes’]

inteligentes

Sugestão de utilização: find_longest_word(words_list)

```python
def find_longest_word(word_list):
  tam = 0;
  palavra = ''
  for word in word_list:
    if len(word) > tam:
      tam = len(word)
      palavra = word
  return palavra
```

```python
lista = ['um', 'tres', 'biblioteca', 'paralamas']
find_longest_word(lista)
```

```python
Marvel_df = filmes_df[filmes_df['Produtora'] == 'Marvel']
DC_df = filmes_df[filmes_df['Produtora'] == 'DC']
MIN_total = min(filmes_df['World'])
MEDIA_total = np.mean(filmes_df['World'])
MAX_total = max(filmes_df['World'])
MIN_Marvel = min(Marvel_df['World'])
MEDIA_Marvel = np.mean(Marvel_df['World'])
MAX_Marvel = max(Marvel_df['World'])
MIN_DC = min(DC_df['World'])
MEDIA_DC = np.mean(DC_df['World'])
MAX_DC = max(DC_df['World'])
```

```python
print('Análise das bilheterias:')
bilheteria = {'produtora' : ["Marvel", 'DC', 'Total'],
              'MAX' : [MAX_Marvel, MAX_DC, MAX_total],
              'MIN' : [MIN_Marvel, MIN_DC, MIN_total],
              'MEDIA' : [MEDIA_Marvel, MEDIA_DC, MEDIA_total]}
pd.DataFrame(bilheteria)
```

```python
# Maior bileteria da Marvel dos últimos 5 filmes (em dólar):
Biggest_Marvel = Marvel_df[Marvel_df['World'] == max(Marvel_df['World'])][['Filme', 'World']]
# Maior bileteria da DC dos últimos 5 filmes (em dólar):
Biggest_DC = DC_df[DC_df['World'] == max(DC_df['World'])][['Filme', 'World']]
```

```python
print('Maior Bilheteria da Marvel e DC, dos últimos 5 filmes, em dólar:')
maiorBilheteria = {'produtora' : ["Marvel", "DC"],
                   'filme' : [Biggest_Marvel['Filme'].values[0], Biggest_DC['Filme'].values[0]],
                   'bilheteria' : [Biggest_Marvel['World'].values[0], Biggest_DC['World'].values[0]]}
pd.DataFrame(maiorBilheteria)
```
